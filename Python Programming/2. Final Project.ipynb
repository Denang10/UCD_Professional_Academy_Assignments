```python
#Creating a class for the board
class Board:
    
    def __init__(self):
        
        #creating our 3x3 grid
        self.cells = [[' ' for _ in range(3)] for _ in range(3)]  
        #Changing the squares to contain numbers rather than spaces to make it easier to read
        for i in range(3):
            for j in range(3):
                if i == 0:
                    self.cells[i][j] = str(j + 1)
                elif i == 1:
                    self.cells[i][j] = str(j + 4)
                else:
                    self.cells[i][j] = str(j + 7)

    #this method is called upon whenever we wish to display the current grid            
    def display(self):
        #Added this as well as the last line for extra style
        print('--+---+--')
        for row in self.cells:
            print(' | '.join(row))
            print('--+---+--')

#Creating the class for the game itself    
class TicTacToe:
    def __init__(self):
        
        #The below lines do the following: instantiate the board, setting player 1's marker, creating a turns attribute, ...
        #... a dictionary that stores the number of turns and another dictionary to keep track of the moves
        self.board = Board()
        self.current_player = 'X'
        self.turns = {'X': 0, 'O': 0}
        self.moves = {'X' : {0 : (0,0)}, 'O' :{0 : (0,0)}}

    #Simple way to switch between players without creating a class for each
    def switch_player(self):
        self.current_player = 'O' if self.current_player == 'X' else 'X'
    
    #This method allows the player to make a move IF the square is free. Afterwards, it stores the move and increases the ...
    #... turns counter for the current player
    def make_move(self, row, col):
        if self.board.cells[row][col] != 'X' and self.board.cells[row][col] != 'O':
            self.board.cells[row][col] = self.current_player
            self.moves[self.current_player][self.turns[self.current_player]] = (row, col)
            self.turns[self.current_player] += 1   
        else:
            print('Invalid move. Please try again.')
    
    #This method gets the input from the user, the user can also input 9 in order to exit out of the game whenever they wish
    def get_input(self):
        
        self.NumberDict = {
            
            '1' : [0,0], '2' : [0,1], '3' : [0,2], '4' : [1,0], '5' : [1,1], \
            '6' : [1,2], '7' : [2,0], '8' : [2,1], '9' : [2,2] 
            
        }
        
        while True:
            try:
                input1 = input('Enter a number between 1 - 9 or press 0 to exit: ')
                
                if input1 == '0':
                    print('Game ended. Thanks for playing!')
                    #There were a few ways one could have ended the running of the code, this is one, another is shown below
                    raise SystemExit            
                
                row = int(self.NumberDict[input1][0])
                col = int(self.NumberDict[input1][1])
                
                if 0 <= row <= 2 and 0 <= col <= 2:
                    return row, col
                else:
                    print('Invalid input. Please enter a number between 1 and 9')
            
            #This exception was included to not allow the user to input anything that is not a number. One could have ...
            #... also set the quitting condition to a string created a custom exception that would allow said string to ...
            #... to be used as an input value but no others.     
            except ValueError:
                print('Invalid input. Please enter a number')
    
    #Checks the rows, columns and diagonals for a winner            
    def check_winner(self):
        for i in range(len(self.board.cells)):
            #Checks rows (uses a functionality of python that was found online)
            if all(self.board.cells[i][j] == self.current_player for j in range(len(self.board.cells))):
                return True

            #Checks columns
            if all(self.board.cells[j][i] == self.current_player for j in range(len(self.board.cells))):
                return True

        #Checks diagonals (uses a \ for line continuation, keeps the code tidier)
        if all(self.board.cells[i][i] == self.current_player for i in range(len(self.board.cells))) or \
            all(self.board.cells[i][len(self.board.cells) - 1 - i] == self.current_player for i in range(len(self.board.cells))):
            return True

        return False
    
    #Since this is ran AFTER the win check, if all squares are filled, it is a draw. It is a bit more useful in the ...
    #... base version of the game but decided to keep it in anyway
    '''def check_tie(self):
        for row in self.board.cells:
            if int in row:
                return False
        return True'''
    
    #Finally we have the method that runs the game, while the code is running and inputs are being required by the user ...
    #... the board gets displayed, input is received and used to make a move and the winner is checked            
    def play_game(self):
        
        while True:
            self.board.display()
            print(f"Player {self.current_player}'s turn")
            row, col = self.get_input()
            self.make_move(row, col)
            
            #This is where the twist occurs! Every time the counter resets to zero for current player, the first ...
            #... move recorded is erased from the board
            if self.turns[self.current_player] > 3:
                self.board.cells[self.moves[self.current_player][self.turns[self.current_player] - 3][0]][self.moves[self.current_player][self.turns[self.current_player] - 3][1]] = ' '
            
            if self.check_winner():
                self.board.display()
                print(f"Player {self.current_player} wins!")
                break
            
            '''if self.check_tie():
                self.board.display()
                print("It's a tie!")
                break'''
            
            #This was added to deal with issues where if a user inputted an incorrect value, it would skip their turn ...
            #... and move onto the next player. This checks to make sure that if it's X's turn and the square chosen ...
            #... is occupied by O, then skip the turn switch. Otherwise switch
            
            if self.current_player == 'X':    
                if self.board.cells[row][col] == 'O':
                    pass
                else:
                    self.switch_player()
            else:
                if self.board.cells[row][col] == 'X':
                    pass
                else:
                    self.switch_player()
                    
start_game = TicTacToe()
start_game.play_game()
#Allows you to play again as many times as you want
while True:
    a = input("do you wanna play again? (y/n)")
    if a.lower() == 'y':
        del start_game
        start_game = TicTacToe()
        start_game.play_game() 
    else:
        break
```
